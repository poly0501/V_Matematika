```python
x=np.linspace(-3,3,200)
plt.plot(x,(np.exp(x)+x-1)/x)
plt.plot(x,x**2-1)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0,7)
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x,y=p
    return (y-x**2+1,np.exp(x)+x-x*y-1)

x1,y1=fsolve(equations,(-2,3))
print (x1,y1)
```
