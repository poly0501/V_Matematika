1.окружность

```python
t = np.arange(0, 2*np.pi, 0.01)
r=4
plt.plot(r*np.sin(t), r*np.cos(t), lw=3)
plt.axis("equal")
plt.show()
```

2.эллипс

```python
u=1.
v=0.5
a=2.
b=1.5
t=np.linspace(0, 2*np.pi,100)
plt.plot(u+a+np.cos(t),v+b*np.sin(t))
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y=np.meshgrid(X,Y)
Z = X +Y
ax.plot_wireframe(X,Y,Z)
X2 = np.arange(-5,5,0.5)
Y2 = np.arange(-5,5,0.5)
X2,Y2=np.meshgrid(X2,Y2)
Z2 = X2 +Y2+5
ax.plot_wireframe(X2,Y2,Z2)
ax.scatter(0,0,0,"z",50,"red")
show()
```

```python
from mpl_toolkits.mplot3d import axes3d
ax = axes3d.Axes3D(plt.figure())
i=np.arange(-1,1,0.01)
X,Y=np.meshgrid(i,i)
Z=X**2-Y**2
ax.plot_wireframe(X,Y,Z,rstride=10,cstride=10)
plt.show()
```

```python
X = 10
N = 50
u=linspace(-X,X,N)
x,y=meshgrid(u,u)
r=sqrt(x**2+y**2)
z=sin(r)/r
fig=figure()
ax=Axes3D(fig)
ax.plot_surface(x,y,z,rstride=1,cstride=1)
```
